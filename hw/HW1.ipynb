```python
import string
import random
import collections
from collections import defaultdict


def read_words_from_file(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()
        words = text.split(', ')
    return words



part1 = '=' * 28
part2 = '||'
part3 = '_' * 16
part4 = '||             |'
part5a = '||           /////'
part5b = '||         (；⌣̀_⌣́)  '
part6a = '||          (＞﹏＜)  '
part6b = '||             |'
part7a = '||      __  (＞﹏＜)  __'
part7b = '||        \____|____/'
part8 = '||             |'
part9a = '||            / \\'
part9b = '||           /   \\'
part9c = '||          /     \\'
part9d = '||         /       \\'
part10 = '||        GAME OVER!!!'
deathmask = '||      __  (X ﹏ X)  __'


def hangman(a):
    if a == 1:
        for i in range(20):
            print(' ')
        print(part1)
    elif a == 2:
        for i in range(20):
            print(part2)
        print(part1)
    elif a == 3:
        print(part3)
        for i in range(20):
            print(part2)
        print(part1)
    elif a == 4:
        print(part3)
        for i in range(3):
            print(part4)
        for i in range(16):
            print(part2)
        print(part1)
    elif a == 5:
        print(part3)
        for i in range(3):
            print(part4)
        print(part5a)
        print(part5b)
        for i in range(13):
            print(part2)
        print(part1)
    elif a == 6:
        print(part3)
        for i in range(3):
            print(part4)
        print(part5a)
        print(part6a)
        print(part6b)
        for i in range(12):
            print(part2)
        print(part1)
    elif a == 7:
        print(part3)
        for i in range(3):
            print(part4)
        print(part5a)
        print(part7a)
        print(part7b)
        for i in range(11):
            print(part2)
        print(part1)
    elif a == 8:
        print(part3)
        for i in range(3):
            print(part4)
        print(part5a)
        print(part7a)
        print(part7b)
        for i in range(3):
            print(part8)
        for i in range(8):
            print(part2)
        print(part1)
    elif a == 9:
        print(part3)
        for i in range(3):
            print(part4)
        print(part5a)
        print(part7a)
        print(part7b)
        for i in range(3):
            print(part8)
        print(part9a)
        print(part9b)
        print(part9c)
        print(part9d)
        for i in range(4):
            print(part2)
        print(part1)
    elif a == 10:
        print(part3)
        for i in range(3):
            print(part4)
        print(part5a)
        print(deathmask)
        print(part7b)
        for i in range(3):
            print(part8)
        print(part9a)
        print(part9b)
        print(part9c)
        print(part9d)
        for i in range(2):
            print(part2)
        print(part10)
        print(part2)
        print(part1)


def theme_choice():
    theme1 = 'города россии'
    theme2 = 'языки мира'
    theme3 = 'лингвистические термины'
    x = ' '
    print('выберите тему')
    print('Города России, Языки мира, Лингвистические термины')
    choice = input()
    # - delete the chosen choice
    choice.lower()
    while x == ' ':
        if choice == theme1:
            x = 'theme1'
        elif choice == theme2:
            print(2)
            x = 'theme2'
        elif choice == theme3:
            print(3)
            x = 'theme3'
        else:
            print('Ну что вы вводите? Введите название темы пожалуйста')
            choice = input()
    return x


def theme_maker(x):
    filename = x + '.txt'
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read()
        words = text.split(', ')
        word = random.choice(words)
    return word


def show(word, guess, showing):
    outcome = ''
    for i in range(len(word)):
        if word[i] == guess:
            outcome += guess
        else:
            outcome += showing[i]
    return outcome

showing = []
def game(word):

    print('У вас есть 10 попыток, чтобы отгадать слово из ', len(word), ' букв')
    print()
    a = 0
    showing = '_ ' * len(word)
    # a stands for attempts used by the player

    while showing != word and a != 10:

        print(showing)
        print('введите букву')
        guess = input()

        if len(guess) != 1:
            print('ну что вы вводите? введите одну букву!')
        else:
            if guess in word:
                print('вы угадали букву')
                showing = show(word, guess, showing)
                hangman(a)
            else:
                print('Вы не угадали!')
                a += 1
                showing = show(word, guess, showing)
                hangman(a)

            print('у вас осталось')
            if 10 - a >= 5 or a == 10:
                print(10 - a, 'попыток')
            elif 10 - a == 1:
                print(10 - a, 'попытка')
            else:
                print(10 - a, 'попытки')
    if a == 10:
        print('кажется, вы проиграли, а слово было - ', word)
        print('сыграете еще разок? да/нет')
        answer = input()
        if answer == 'да':
            main()
        else:
            print('гудбай')
    
    
    elif showing == word:
        print('Ура! Вы выиграли')
        print('сыграете еще разок? да/нет')
        answer = input()
        if answer == 'да':
            main()
        else:
            print('гудбай')
        


def main():
    x = theme_choice()
    word = theme_maker(x)
    game(word)




if __name__ == '__main__':
    main()
```
